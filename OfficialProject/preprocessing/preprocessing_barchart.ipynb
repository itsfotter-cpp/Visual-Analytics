```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "/content/drive/MyDrive/VisualAnalytics/Formatted2022FIFA.xlsx"
```

```python
df = pd.read_excel(file)
```

```python
df["Height"][1]
```

```python
df["Weight"][1]
```

```python
for idx, d in enumerate(df["Height"]):
  df["Height"][idx] = int(df["Height"][idx].replace("cm", ""))
```

```python
for idx, d in enumerate(df["Weight"]):
  df["Weight"][idx] = int(df["Weight"][idx].replace("kg", ""))
```

```python
df[["Height", "Weight"]].head(5)
```

```python
dfHeight = df.groupby("Best Position", as_index=False)[["Height", "Weight"]].mean()
```

```python
dfHeight
```

```python
ruoli = ['GK', "CB", "LB", "RB", "LWB", "RWB", "CM", "CDM", "CAM", "LM", "RM", "LW", "RW", "CF", "ST"]

dfHeight["Best Position"] = pd.Categorical(dfHeight['Best Position'], ruoli)
```

```python
dfHeight = dfHeight.sort_values("Best Position")
```

```python
dfHeight
```

```python
dfHeight.index.name = 'ID'
```
